### 1) White noise realisations
The properties of a *white noise* are : <br>
* All variables are independant and independantly distributed (iid)
* Zero mean with finite variance
* Each value has a zero correlation with the other values in the time series

```python
from random import seed
from random import gauss
from pandas import Series
from pandas.plotting import autocorrelation_plot
import matplotlib.pyplot as plt
import numpy as np
from Utils import *
%load_ext autoreload
%autoreload 2
plt.style.use("default")
plt.rcParams['font.size'] = 13
plt.rcParams['axes.linewidth'] = 1.5

# Generate 3 white noise sequences
N = 200000
t = np.arange(0,N,1).T
series = np.empty([3,N])
for i in range(3):
    seed(i) # Seed for the generated random number
    series[i,:] = Series([gauss(0.0, 1.0) for i in range(N)])

# Plots and stats summary
fig, ax = plt.subplots(3, 1, 
                       constrained_layout = True, figsize=(20,5))
for i in range(3):
    ax[i].plot(t, series[i,:], label="Series " + str(i+1))
    ax[i].set_xlabel('Time [s]'), ax[i].set_ylabel('Amplitude [-]')
    ax[i].grid()
    ax[i].legend()
    print('Mean of time series ',i,' :',np.mean(series[i,:]),' variance of time series ',i,' :',np.var(series[i,:]))
```

### 2) Random walk
The properties of a random walk are :
* For $Z_1, \ldots, Z_N$ being $\pm 1$ with a certain probability $P_i$, one can construct the following time series : $x_{i+1} = x_i + Z_{i+1}$
* Now, this can be extended with any number (not just $\pm1$). To do that, we use the randomly generated sequences borrowed from *1)* and simply perform : $x_{i+1} = x_i + S_{i+1}$ with $S$ being our randomly generated sequence.

```python
from Utils import *
%load_ext autoreload
%autoreload 2

rndWalks = randomWalk(series)

fig, ax = plt.subplots(3, 1, 
                       constrained_layout = True, figsize=(20,5))
for i in range(3):
    ax[i].plot(t, rndWalks[i,:], label="Random walk "+str(i+1))
    ax[i].set_xlabel('Time [s]'), ax[i].set_ylabel('Amplitude [-]')
    ax[i].grid()
    ax[i].legend()
```

### 3) 1$^{st}$ order Gauss-Markov process
This is a correlated random process. the time-series is of the form : $x_{t+1} = \eta x_{t} + (1-\eta)u_t = \eta^t u_0 + (1- \eta) \sum_{i=1}^{t} \eta^{t-i}u_i$. Now, $0 < \eta < 1$ is related to the correlation number $\tau_{exp}$ as follows : $\eta = e^{\frac{-1}{\tau_{exp}}}$. Hence, $\tau_{exp}$ defines how much delayed the autocorrelation of the time series is, i.e. $Corr(x_t,x_{t+k}) = \eta^k$.

```python
from Utils import *
%load_ext autoreload
%autoreload 2

tau = 2000 # Correlation time
Gauss_Markov_Processes, eta = gaussMarkov(series, tau)

fig, ax = plt.subplots(3, 1, 
                       constrained_layout = True, figsize=(20,5))
for i in range(3):
    ax[i].plot(t, Gauss_Markov_Processes[i,:], label="Gauss-Markov process "+str(i+1))
    ax[i].set_xlabel('Time [s]'), ax[i].set_ylabel('Amplitude [-]')
    ax[i].grid()
    ax[i].legend()
fig.suptitle(r'1$^{st}$ order Gauss-Markov processes with $\tau_{exp} = $'+str(tau)+' and $\eta = $'+str(eta))
```
